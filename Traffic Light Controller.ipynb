```python
import time
import os

# Clear the console
def clear_console():
    os.system('cls' if os.name == 'nt' else 'clear')

# Print signal status with colored emojis
def display_signals(active_road, timer):
    signals = ['🔴'] * 4  # Red signals by default
    signals[active_road] = '🟢'  # Set the active signal to green

    # Console layout
    print("\n============ TRAFFIC SIGNAL CONTROLLER ============\n")
    print(f"⏱️ Timer: {timer} seconds remaining\n")
    print("Signal Status:")
    print(f" Road 1: {signals[0]}")
    print(f" Road 2: {signals[1]}")
    print(f" Road 3: {signals[2]}")
    print(f" Road 4: {signals[3]}\n")

    print("Intersection View:\n")
    print(f"         Road 1 {signals[0]}")
    print("              |")
    print("              |")
    print("    -------------------------")
    print(f"Road 4 {signals[3]}       Road 2 {signals[1]}")
    print("    -------------------------")
    print("              |")
    print("              |")
    print(f"         Road 3 {signals[2]}")
    print("\n===================================================\n")

# Controller loop
def traffic_signal_controller():
    while True:
        for road in range(4):
            for timer in range(30, 0, -1):
                clear_console()
                display_signals(road, timer)
                time.sleep(1)

# Run
traffic_signal_controller()
```
